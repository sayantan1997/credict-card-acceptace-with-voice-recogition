```python

#IMPORTING DATASET
import pandas as pd
df=pd.read_csv("/home/sayantan/Desktop/Worksheet in Project Credit Card.csv")
#DISPLAYING THE DATASET
df
```

```python
#DISPLAYING THE PROPERTIES OF THE DATASET
df.describe()
```

```python
#REPLACING NO WITH 0'S AND YES WITH 1'S FOR COLUMN 'CARD'
df.card.replace("yes",1,inplace=True)
df.card.replace("no",0,inplace=True)
df
```

```python
##REPLACING NO WITH 0'S AND YES WITH 1'S FOR COLUMN 'OWNER' , 'SELFEMP'
df.owner.replace("yes",1,inplace=True)
df.owner.replace("no",0,inplace=True)
df.selfemp.replace("yes",1,inplace=True)
df.selfemp.replace("no",0,inplace=True)
df
```

```python
#CHECKING FOR NULL VALUES ACROSS THE DATASET
df[df.isnull()==True].count()
```

```python
#FINDING THE CO-RELATION
print("CO-RELATION BETWEEN CARD AND REPORTS :",df["card"].corr(df["reports"]))
print("CO-RELATION BETWEEN CARD AND AGE :",df["card"].corr(df["age"]))
print("CO-RELATION BETWEEN CARD AND INCOME :",df["card"].corr(df["income"]))
print("CO-RELATION BETWEEN CARD AND SHARE :",df["card"].corr(df["share"]))
print("CO-RELATION BETWEEN CARD AND EXPENDITURE :",df["card"].corr(df["expenditure"]))
print("CO-RELATION BETWEEN CARD AND OWNER :",df["card"].corr(df["owner"]))
print("CO-RELATION BETWEEN CARD AND SELF-EMPLOYEED :",df["card"].corr(df["selfemp"]))
print("CO-RELATION BETWEEN CARD AND DEPENDENT'S :",df["card"].corr(df["dependents"]))
print("CO-RELATION BETWEEN CARD AND MONTH'S OF RESIDING :",df["card"].corr(df["months"]))
print("CO-RELATION BETWEEN CARD AND MAJOR CARD :",df["card"].corr(df["majorcards"]))
print("CO-RELATION BETWEEN CARD AND ACTIVE :",df["card"].corr(df["active"]))
```

```python
#SELECTING FEATURES FOR TRAINING
l=dict()
for i in df.columns[1:]:
    l[str(i)]=df["card"].corr(df[str(i)])
k=sorted(l.values(),reverse=True)
k1=k[0:5]
features=list()
for j in l.keys():
    if l[j] in k1:
        features.append(j)
print("FEATURES CHOSEN FOR TRAINING (BASED ON CO-RELATION) :",features)
```

```python
#age outlier
print("AGE OUTLIER'S :    ")
print(df[(df["age"]-df["age"].mean())<-2*df["age"].std()])
print(len(df[(df["age"]-df["age"].mean())<-2*df["age"].std()]))

print("REMOVING AGE OUTLIER'S ...........  ")
df.drop(df[df["age"]-df["age"].mean()<-2*df["age"].std()].index,inplace=True)
print(len(df))
print("DONE")
```

```python
#DECISION TREE
from sklearn.model_selection import train_test_split
train,test=train_test_split(df,test_size=0.4,random_state=176)

print("length of modified dataset :  ",len(df))
print("length of train dataset :  ",len(train))
print("length of test dataset :  ",len(test))
```

```python
#SETTING UP DECISION TREE CLASSIFIER

from sklearn.metrics import accuracy_score
from sklearn.tree import DecisionTreeClassifier as dtc
import numpy as np
c1=dtc(criterion="gini",random_state=200,min_samples_leaf=5,max_depth=5)
outcomes=np.array(train["card"])
c1.fit(train[features],outcomes)

#TESTING THE DATASET

k=c1.predict(test[features])


print(accuracy_score(test["card"],k)*100)

print(c1.predict_proba(test[features]))
```

```python
#SETTING UP RANDOM FOREST CLASSIFIER

from sklearn.ensemble import RandomForestClassifier as rfc

c2=rfc(n_jobs=2,n_estimators=200,random_state=200,criterion="gini")
c2.fit(train[features],outcomes)

#TESTING THE DATASET

k1=c2.predict(test[features])

print("accuracy : ",accuracy_score(test["card"],k1)*100)

print("probabilities of acceptance or decline ")

print(c2.predict_proba(test[features]))
```

```python
#SETTING UP LOGISTIC REGRESSION CLASSIFIER

from sklearn.linear_model import LogisticRegression as lrc

c3=lrc(penalty='l2',C=0.01,fit_intercept=1.0)

c3.fit(train[features],outcomes)

k2=c3.predict(test[features])

print("accuracy : ",accuracy_score(test["card"],k2)*100)

print(c3.predict_proba(test[features]))
```

```python
#SETTING UP NAIVE BIAS CLASSIFIER

from sklearn.naive_bayes import GaussianNB as gnb


count1=(df['card'][df['card']==1]).count()
count2=(df['card'][df['card']==0]).count()
totalcount=df['card'].count()
print("positive examples count : ",count1)
print("negetive examples count : ",count2)
print("total count : ",totalcount)

p1=count1/totalcount
p2=count2/totalcount
print("priors of class's : ",p1," ",p2)
ptotal=np.array([[p1],[p2]])


c4=gnb(ptotal)

c4.fit(train[features],outcomes)

#prediction

k3=c4.predict(test[features])

print("accuracy : ",accuracy_score(test['card'],k3)*100)

print(c4.predict_proba(test[features]))
```
